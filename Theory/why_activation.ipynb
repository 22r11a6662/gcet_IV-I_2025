```python
# Re-run after kernel reset
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

# Generate input data
X = np.linspace(0, 2 * np.pi, 200).reshape(-1, 1)
y = np.sin(X).ravel()

# Models
models = {
    'No Activation (Linear)': MLPRegressor(hidden_layer_sizes=(100,), activation='identity', max_iter=5000, random_state=42),
    'ReLU Activation': MLPRegressor(hidden_layer_sizes=(100,), activation='relu', max_iter=5000, random_state=42),
    'Tanh Activation': MLPRegressor(hidden_layer_sizes=(100,), activation='tanh', max_iter=5000, random_state=42)
}

# Plot setup
plt.figure(figsize=(15, 8))

# Train and predict with each model
for i, (name, model) in enumerate(models.items(), start=1):
    model.fit(X, y)
    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)

    # Plot predictions
    plt.subplot(1, 3, i)
    plt.plot(X, y, label="True $\sin(x)$", linewidth=2)
    plt.plot(X, y_pred, label=f"{name} (MSE={mse:.4f})", linestyle='--')
    plt.title(name)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.legend()

plt.suptitle("Neural Network Approximations of $y = \sin(x)$ with Different Activation Functions", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function and its gradient
def f(x, y):
    return (x - 3)**2 + (y - 2)**2 + 1

def grad_f(x, y):
    return np.array([2 * (x - 3), 2 * (y - 2)])

# Initialize
x0, y0 = 8.0, -4.0  # starting point
learning_rate = 0.1
iterations = 30
path = [(x0, y0)]
x, y = x0, y0

# Perform gradient descent
for _ in range(iterations):
    grad = grad_f(x, y)
    x -= learning_rate * grad[0]
    y -= learning_rate * grad[1]
    path.append((x, y))
path = np.array(path)

# Create the surface
x_range = np.linspace(-1, 9, 100)
y_range = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = f(X, Y)

# Plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6, edgecolor='none')
ax.plot(path[:, 0], path[:, 1], f(path[:, 0], path[:, 1]), color='red', marker='o', label='Gradient Descent Path')
ax.set_title("3D Gradient Descent on $f(x, y) = (x - 3)^2 + (y - 2)^2 + 1$")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.legend()
plt.tight_layout()
plt.show()
```
